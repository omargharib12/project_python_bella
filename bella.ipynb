```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('dailyActivity_merged.csv')
```

```python
df
```

```python
df.Id.nunique()
```

```python
cols = ['Id','ActivityDate','TotalSteps','VeryActiveMinutes','FairlyActiveMinutes','LightlyActiveMinutes','SedentaryMinutes','Calories']
df =df[cols]
```

```python
df
```

```python
df.rename(columns={'ActivityDate':'date'}, inplace=True)
```

```python
df
```

```python
df['TotalMinutes'] = df.VeryActiveMinutes + df.FairlyActiveMinutes + df.LightlyActiveMinutes + df.SedentaryMinutes
```

```python
df
```

```python
df['TotalHours'] = round(df.TotalMinutes / 60)
```

```python
df
```

```python
df.info()
```

```python
df.date = pd.to_datetime(df.date)
```

```python
import datetime as dt
df['DayOfWeek'] = df.date.dt.day_name()
```

```python
plt.figure(figsize=((6,4)))
plt.hist(df.DayOfWeek, bins=7, color = 'lightskyblue' , width = (0.6))
plt.grid(True)
plt.xlabel('DayOfweek')
plt.ylabel('Freq')
plt.title('Freq By Day Of Week')
plt.show('')
```

```python
plt.figure(figsize = ((8,6)))
plt.scatter(df.TotalSteps , df.Calories , c =df.Calories)

median_steps = 7405
median_Caloreis = 2134
plt.axhline(median_Caloreis, color = 'blue', label ='median_Calorei' )
plt.axvline(median_steps, color = 'red', label ='median_steps' )

plt.xlabel('steps')
plt.ylabel('Calories')
plt.title('Calories by total steps')
plt.legend()
plt.show()
```

```python
plt.figure(figsize = ((8,6)))
plt.scatter(df.TotalHours , df.Calories , c =df.Calories)

median_Hours = 24
median_Caloreis = 2134
plt.axhline(median_Caloreis, color = 'blue', label ='median_Calorei' )
plt.axvline(median_Hours, color = 'red', label ='median_Hours' )

plt.xlabel('Hours')
plt.ylabel('Calories')
plt.title('Calories by total Hours')
plt.legend()
plt.show()
```
